```python
import json
```

```python
def load_data(filename):
    with open(filename,'r') as f:
        data = json.load(f)
    return data
```

```python
data = load_data('data.json')
```

```python
def display_users(data):
    print('user data pages\n')
    for user in data['users']:
        print(f"ID{user['id']}: users name: {user['name']} - users friends: {user['friends']} - like page: {user['liked_pages']}")
    print('\n users like pages\n')
    for page in data['pages']:
        print(f"ID{page['id']}: {page['name']}")

data = load_data('data.json')
display_users(data)
```
